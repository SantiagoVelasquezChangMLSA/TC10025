```python

#1.
print("1. Cuál es el vino con mayor cantidad de alcohol?")
print(df[df['Alcohol']==df.Alcohol.max()])
print("---------------------------------------------------------------")

#2.
print("2. ¿Qué vinos superan los 13.5 grados de alcohol?")
print(df[df.Alcohol>13.5])
print("---------------------------------------------------------------")

#3.
print("3. EL OD280 se usa para medir la calidad del vino, ¿qué vinos están en el top 10% de esta medida?")
percentil_90 = df['OD280'].quantile(0.90)
top_vinos = df[df['OD280'] > percentil_90]
print(top_vinos)
print("---------------------------------------------------------------")

#4.
print("¿Cuántos vinos tienen un contenido de magnesio superior al promedio?")
print(df[df['Magnesium'] > df['Magnesium'].mean()].shape[0])
print("---------------------------------------------------------------")

#5.
print("¿Cuál es la relación entre alcohol y magnesio?")
plt.scatter(df['Total_Phenols'], df['Flavanoids'])
plt.title('Relación entre Fenoles Totales y Flavonoides')
plt.xlabel('Total_Phenols')
plt.ylabel('Flavanoids')
plt.show()
correlacion = df['Total_Phenols'].corr(df['Flavanoids'])
print(f"Coeficiente de correlación: {correlacion}")
print("Relación muy fuerte")
print("---------------------------------------------------------------")
```
