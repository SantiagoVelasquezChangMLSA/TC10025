```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("wine-clustering.csv")
```

```python
# Create a heatmap
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

plt.title('Correlation Heatmap')
plt.show()
```

```python
for column in df.columns:
    plt.figure()  # Create a new figure for each histogram
    plt.hist(df[column], bins=10)  # You can adjust the number of bins as needed
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.show()
```

```python
# Get the list of column names
columns = df.columns
num_cols = len(columns)
num_rows = (num_cols + 1) // 2  # Calculate the number of rows needed

# Create subplots with two graphs per row
fig, axes = plt.subplots(num_rows, 2, figsize=(12, num_rows * 4))

# Flatten the 2D array of axes to make it easier to iterate
axes = axes.flatten()

# Create a boxplot for each variable
for i, column in enumerate(columns):
    ax = axes[i]
    sns.boxplot(x=df[column], palette="Set2", ax=ax)
    ax.set_title(f'Boxplot for {column}')
    ax.set_xlabel('Values')

# Hide any unused subplots
for i in range(num_cols, len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
```
